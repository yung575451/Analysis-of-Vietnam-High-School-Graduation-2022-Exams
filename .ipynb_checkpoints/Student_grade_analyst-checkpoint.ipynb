```python
import pandas as pd
import numpy as np
from unidecode import unidecode
from statsmodels.tsa.arima.model import ARIMA
import warnings
from scipy import stats
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
# Filter out the warning messages
warnings.filterwarnings('ignore')
```

# Preprocess Data

```python
df=pd.read_csv('diemthi2022.csv')

df = df.rename(columns={'Foreign Language': 'English'})
df = df.rename(columns={'GeographyCivic education': 'Geography'})
df = df.rename(columns={'Civic education': 'Civic_education'})
df = df.rename(columns={'GPA of Social Science': 'SocialS_GPA'})
df = df.rename(columns={'GPA of Natural Science': 'NaturalS_GPA'})

# Preprocess the 'Province/City' column
df['Province/City'] = df['Province/City'].apply(lambda x: unidecode(str(x).lower()))
```

```python
df.to_csv('StudentGrade_2022.csv',index=False)
```

```python
city_population=pd.read_csv('population.csv')

# Replace commas with an empty string
city_population = city_population.replace(',', '', regex=True)
population=city_population[['Population_1999','Population_2009','Population_2019']]
city_population['Predict_population_2020']=None
city_population['Predict_population_2021']=None
city_population['Predict_population_2022']=None
```

```python
for index, row in population.iterrows():
    population_1999 = row['Population_1999']
    population_2009 = row['Population_2009']
    population_2019 = row['Population_2019']
    
    # Assuming you have the population data in a dictionary
    data = {'Year': [1999, 2009, 2019], 'Population': [population_1999, population_2009, population_2019]}

    dfrow = pd.DataFrame(data)
    dfrow.set_index('Year', inplace=True)

    # Convert 'Population' column to numeric
    dfrow['Population'] = pd.to_numeric(dfrow['Population'], errors='coerce')

    # Use linear interpolation to fill the years between
    dfrow = dfrow.reindex(np.arange(1999, 2020))
    dfrow = dfrow.interpolate(method='linear')

    dfrow['Population']=dfrow['Population'].round()
    # Fit the ARIMA model
    model = ARIMA(dfrow['Population'], order=(2,0,0))  # Modify the order parameters as needed
    model_fit = model.fit()

    # Make the prediction. The second element of the returned tuple is a standard error
    prediction_2020 = model_fit.forecast(steps=1).item()  # 2020, 2021, 2022
    prediction_2021 = model_fit.forecast(steps=2).values[1].item()   # 2020, 2021, 2022
    prediction_2022 = model_fit.forecast(steps=3).values[2].item()  # 2020, 2021, 2022

    # Convert the prediction to an integer and then round it
    predicted_population_2020 = int(round(prediction_2020))
    predicted_population_2021 = int(round(prediction_2021))
    predicted_population_2022 = int(round(prediction_2022))
    city_population['Predict_population_2020'][index]=predicted_population_2020
    city_population['Predict_population_2021'][index]=predicted_population_2021
    city_population['Predict_population_2022'][index]=predicted_population_2022
```

```python
city_population['Name']=city_population['Name'].apply(lambda x: unidecode(str(x).lower()))
city_population=city_population[['Name','Predict_population_2020','Predict_population_2021','Predict_population_2022']]
city_population = city_population.rename(columns={'Name': 'Location'})
city_population.loc[city_population['Location'] == 'ha noi (incl. ha tay) [hanoi]', 'Location'] = 'ha noi'
city_population.loc[city_population['Location'] == 'thanh pho ho chi minh [ho chi minh city]', 'Location'] = 'thanh pho ho chi minh'
```

```python
cities_location=pd.read_csv('cities_location.csv', encoding='latin1')
cities_location['Location']=cities_location['Location'].apply(lambda x: unidecode(str(x).lower()))
cities_location['Latitude']=cities_location['Latitude'].replace('N', '', regex=True)
cities_location['Longitude']=cities_location['Longitude'].replace('E', '', regex=True)
```

```python
cities_location
```

```python
for index, row in cities_location.iterrows():
    Latitude = row['Latitude']
    Longitude = row['Longitude']
    
    degree_la= float(Latitude[:Latitude.find('°')])
    
    decimalDegrees=degree_la + minutes_la/60
    
    cities_location['Latitude'][index]=decimalDegrees
    
    degree_lo = float(Longitude[:Longitude.find('°')])
    
    decimalDegrees=degree_lo + minutes_lo/60
    
    cities_location['Longitude'][index]=decimalDegrees
```

```python
# Left join
result = city_population.merge(cities_location, how='left', on='Location')


result['Longitude']=result['Longitude']-1
result['Latitude']=result['Latitude']-1
```

## Rural student English is not as good as Urban student ?

```python
df=pd.read_csv('StudentGrade_2022.csv')
city_information=pd.read_csv('cities_infomation.csv')
urban=city_information[city_information['Rank']==1]
rural=city_information[city_information['Rank']==2]
df = df.rename(columns={'Province/City': 'Location'})
df = df[df['English'].notna()]

urban=urban['Location']
rural=rural['Location']
urban_student = df.merge(urban, how='inner', on='Location')
rural_student = df.merge(rural, how='inner', on='Location')
```

```python
urban_grades = urban_student['English']
rural_grades = rural_student['English']

t_statistic, p_value = stats.ttest_ind(urban_grades, rural_grades)
```

```python
# Calculate the mean and standard deviation
mean1 = np.mean(urban_grades)
std1 = np.std(urban_grades)
mean2 = np.mean(rural_grades)
std2 = np.std(rural_grades)

# Plotting
x = np.arange(2)  # x-axis positions for the bars
labels = ['Urban', 'Rural']
means = [mean1, mean2]
stds = [std1, std2]

plt.bar(x, means, tick_label=labels, yerr=stds, alpha=0.7)
plt.xlabel('Arrays')
plt.ylabel('Mean')
plt.title('Comparison of Means with Standard Deviation')
plt.show()
```

```python
# Round the data to nearest 0.2
#rounded_data1 = np.round(urban_grades * 2) / 2
#rounded_data2 = np.round(rural_grades * 2) / 2

rounded_data1 = urban_grades
rounded_data2 = rural_grades

# Create bins with step of 0.2
bins = np.arange(min(min(rounded_data1), min(rounded_data2)), max(max(rounded_data1), max(rounded_data2)) + 0.5, 0.5)

# Calculate histograms
hist1, edges1 = np.histogram(rounded_data1, bins=bins, density=True)
hist2, edges2 = np.histogram(rounded_data2, bins=bins, density=True)

# Convert histogram values to percentages
hist_percentage1 = hist1 / sum(hist1) * 100
hist_percentage2 = hist2 / sum(hist2) * 100

# Plot
plt.figure(figsize=(10, 5))
plt.plot(edges1[:-1], hist_percentage1, marker='o', label='Urban Student')
plt.plot(edges2[:-1], hist_percentage2, marker='o', label='Rural Student')
plt.title('Percentage Distribution of Point')
plt.xlabel('Student Point')
plt.ylabel('Percentage (%)')
plt.xticks(bins)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

df=pd.read_csv('StudentGrade_2022.csv')
city_information=pd.read_csv('cities_infomation.csv')
df = df.rename(columns={'Province/City': 'Location'})
```

```python
city_major=city_information[['Location']]
city_major['total_sience']=None
city_major['total_social']=None
city_major['major']=None
```

```python
df
```

```python
# Add points to the map
for idx, row in city_major.iterrows():
    location=row['Location']
    major_science=df[(df['Location']==location)&(df['Biological'].notna())]
    major_social=df[(df['Location']==location)&(df['Civic_education'].notna())]
    major=None
    science_row_count = len(major_science)
    social_row_count = len(major_social)
    
    if science_row_count>social_row_count:
        major=1
    else:
        major=2
        
    city_major['total_sience'][idx]=science_row_count
    city_major['total_social'][idx]=social_row_count
    city_major['major'][idx]=major
```

```python
city_major = city_major.merge(city_information, how='inner', on='Location')
```

```python
# Create a folium map centered around the mean of your coordinates
m = folium.Map(location=[city_major['Latitude'].mean(), city_major['Longitude'].mean()], zoom_start=6)

# Add points to the map
for idx, row in city_major.iterrows():
    if row['major']==2:
        color = 'blue'
    else:
        color = 'red'
    
    folium.Marker([row['Latitude'], row['Longitude']],
                 icon=folium.Icon( color=color,icon='circle', prefix='fa'),  # Use 'circle' as the marker icon
                icon_size=(2, 2)  # Adjust the icon size as needed
                 ).add_to(m)
    
# Create a GeoJSON layer for Vietnam
highlighted_country = folium.GeoJson(
    data=open('vietnam-with-regions_.geojson', encoding='utf-8').read(),
    style_function=lambda x: {'fillColor': 'black', 'color': 'black'}
)

# Add the GeoJSON layer to the map
highlighted_country.add_to(m)

# Create a legend
legend_html = '''
    <div style="position: fixed; bottom: 50px; left: 50px; width: 300px; height: 160px; 
                border:2px solid grey; z-index:9999; font-size:28px;
                background-color: white;
                opacity: 0.9;
                padding: 10px;">
        <p><i class="fa fa-circle" style="color:red"></i> Science Majority</p>
        <p><i class="fa fa-circle" style="color:blue"></i> Social Majority</p>
    </div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# Display the map
m
```
